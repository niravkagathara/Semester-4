```python
###morst imp imp 
x = range(1,11,1)
cxMarks= [8,9,6,3,5,7,4,1,2,6]
cyMarks= [5,8,9,6,3,2,4,8,8,9]


# write a code to generate below graph
plt.plot(x,cxMarks,color='red',ls='--',marker='o')
plt.plot(x,cyMarks,color='y',ls=':',marker='>')
plt.xlabel('Roll number')
plt.ylabel('SPI')
plt.annotate(xy=[8,1],text='Lowest SPI')
plt.annotate(xy=[1,9],text='high SPI')

plt.legend(['cx','cy'],loc=3)
plt.show()
```

### 01) WAP to demonstrate the use of Pie chart.

```python
#most imp
x = [10,20,30,40]
y = ["food","travel","accomodation","shopping"]
c = ['r','g','b','y']
ex = [0.5,0,0.0,0.0]
plt.pie(x,labels=y,colors=c,explode=ex,shadow=True,autopct="%0.1f%%",radius=1.2,labeldistance=1.2)
plt.title("Expenses")
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x = ["python","c","c++","java"]
y = [85,70,60,80]
c = ['r','g','b','y']
plt.bar(x,y,color=c,edgecolor='b',alpha=0.5)
plt.title("Popularity of language")
plt.xlabel("Language")
plt.ylabel("Popularity")
def addlabels(x,y):
    for i in range (len(x)):
        plt.text(i,y[i],y[i])
addlabels(x,y)
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
c = ['r','g','b','y','black']
si = [100,200,300,400,500]
plt.scatter(x,y,color=c,marker='^',s=si)
plt.title("Scatter plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
# plt.show()
plt.savefig("MyChart.jpg")
```
